# Stack Plot

```python
days = [1,2,3,4,5];sleeping = [7,8,6,11,7];eating = [2,3,4,3,2];working = [7,8,7,2,2];playing = [8,5,7,8,13]
plt.plot([],[], color='m', label='Sleeping')
plt.plot([],[], color='c', label='Eating')
plt.plot([],[], color='r', label='Working')
plt.plot([],[], color='k', label='Playing')
plt.stackplot(days, sleeping, eating, working ,playing, colors=['m','c', 'r', 'k'])
plt.xlabel('days');plt.ylabel('Activities')
plt.title('Persons Weekly Spent Time per activities\nSpring 2018')
plt.legend();plt.show()
```

# Case Study 2: Analyzing Gun Deaths in the United States (20122014)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='white', color_codes=True)
%matplotlib inline
```

```python
dataset = pd.read_csv('H:/study/5th_semester/DataAnylasis/Assignmnet_Data_DataAnalysis_2/dataset/guns-data-master/full_data.csv', index_col=0)
print(dataset.shape)
dataset.index.name = 'Index'
dataset.columns = map(str.capitalize, dataset.columns)
dataset.head(5)
```

```python
dataset_Gun = dataset
dataset_Gun.sort_values(['Year', 'Month'],inplace=True)
```

```python
dataset_Gun.Sex.value_counts(normalize=False)
```

```python
dataset_byGender = dataset_Gun.groupby('Sex').count()
dataset_byGender
```

```python
dataset_suicide_Gender =dataset_Gun[dataset_Gun["Intent"] =="Suicide"]
dataset_suicide_Gender.Sex.value_counts(normalize=False).plot.bar(title='Annual U.S.\\suicide gun deaths \n 2012-2014, by gender')
```

```python
dataset_byGender.plot.bar(title='Annual U.S. suicide gun deaths \n 2012-2014, by gender')
```

```python
dataset_byRace = dataset_Gun.Race.value_counts(ascending=False)*100/100000
dataset_byRace
```

```python
(dataset_Gun.Race.value_counts(ascending=False)*100/100000).plot.bar(title='Percent death toll from guns inthe United States \nfrom 2012 to 2014, by race')
```

```python
dataset_Gun.Intent.value_counts(sort =True,ascending=False)
```

```python
dataset_Gun.Intent.value_counts(sort=True).plot.bar(title='Annual number of gun deaths in the United States on average \n from 2012 to 2014, by cause')
```

```python
dataset_Gun.Intent.value_counts(ascending=False)*100/100000
```

```python
(dataset_Gun.Intent.value_counts(ascending=False)*100/100000).plot.bar(title='Rate gun deaths in the U.S. per 100,000 population \n2012-2014, by race')
```

```python
dataset_suicide=dataset[ dataset["Intent"]=="Suicide"]
datasetSuicide= dataset_suicide.Year.value_counts(ascending=False) *100/100000
datasetSuicide.sort_values(ascending=True)
```

```python
datasetSuicide.sort_values(ascending=True).plot.bar(title='Percentage of annual suicide gun deaths in the United States \nfrom 2012 to 2014, by year')
```
