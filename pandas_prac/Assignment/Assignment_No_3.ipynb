# Exporting Multiple DataFrames to excel

```python
Names = ['BOB','Jhon','Mel']
Grades = [89,98,78]
bsdegrees = [1,1,0]
msdegrees = [1,0,0]
phddegrees = [1,0,0]
degress = list(zip(Names,Grades,bsdegrees,msdegrees,phddegrees))
Colm = ['Names','Grades','BS','MS','PhD']

_Names = ['Usman','Noman','Nadeem']
_Grades = [89,98,78]
_bsdegrees = [1,1,0]
_msdegrees = [1,0,0]
_phddegrees = [1,0,0]
_degress = list(zip(_Names,_Grades,_bsdegrees,_msdegrees,_phddegrees))
_Colms = ['Names','Grades','BS','MS','PhD']

df1 = pd.DataFrame(data=degress,columns=Colm)
df2 = pd.DataFrame(data=_degress,columns=_Colms)

write = pd.ExcelWriter("H:/study/5th_semester/DataAnylasis/Assignmnet_Data_DataAnalysis_2/Assignmnet_two_data/dataset/dataframes.xlsx",engine="xlsxwriter")

df1.to_excel(write,sheet_name='Sheet_1')
df2.to_excel(write,sheet_name='Sheet_2')

write.save()
```
