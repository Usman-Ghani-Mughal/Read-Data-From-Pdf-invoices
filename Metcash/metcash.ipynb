```python
def isPresent(df,str):
    result_ser = df.iloc[:,0].str.contains(str)
    result_list = result_ser.tolist()
    for i in result_list:
        if i == True:
            return True
    return False
```

```python
def findTableStartIndex(df,starting_index):
    start_index = starting_index
    result_list = df.iloc[start_index:,0].str.contains('NaN').tolist()
    for item in result_list:
        if item != False:
            start_index = start_index + 1
        else:
            return start_index
    return start_index
```

```python
def findEndingIndex(df,starting_index):
    end_index = starting_index
    result_list = df.iloc[end_index:,0].str.contains('NaN').tolist()
    for item in result_list:
        if item == False:
            end_index = end_index + 1
        else:
            return end_index
    return end_index
```

```python
def findColumIndex(df,strs):
    #df.columns[df.isin(['Line'])]
    result_list = []
    num_of_col = df.shape[1]
    for col_index in range(0,num_of_col):
        result_list=df.iloc[:,col_index].astype(str).str.contains(strs).tolist()
        for value in result_list:
            if value == True:
                return col_index
    return -1
```
