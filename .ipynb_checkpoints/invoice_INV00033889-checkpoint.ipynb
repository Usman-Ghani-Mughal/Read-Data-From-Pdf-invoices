```python
from tabula import read_pdf
import pandas as pd
```

```python
filename = 'INV00033889'
df = read_pdf(filename+'.pdf')
df
```

```python
table_df=df[3:25]
table_df
```

```python
Qty = []
item_code = []
description = []
unit_price_GST = []
total_GST = []


Qty = table_df['Unnamed: 0'].tolist()
item_code = table_df['Unnamed: 1'].tolist()
description = table_df['Unnamed: 2'].tolist()
unit_price_GST = table_df['(02) 9635 7033'].tolist()
total_GST = table_df['Unnamed: 5'].tolist()


data_dic = {'QTY':Qty,'Item Code':item_code,'Description':description,"Unit Price(inc-GST)":unit_price_GST,
            "Total(inc-GST)":total_GST}

dataframe = pd.DataFrame(data=data_dic)
dataframe.to_csv(filename+'.csv',index=None,header=True)
```
