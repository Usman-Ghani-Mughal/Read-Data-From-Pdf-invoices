```python
import tabula
import pandas as pd
# import camelot
```

```python
filename = 'Milk_3'
df = tabula.read_pdf(filename+'.pdf')
df.head(13)
```

```python
#print(df['Unnamed: 0'].str.find('Delivery Note'))

df['Unnamed: 0'].str.contains('Delivery Note')
```

```python
table_df=df[12:28]
table_df
```

```python
table_df = table_df.dropna(axis='columns')
table_df
```

```python

product_description = [] #
product_code = [] #
quantity = [] #
list_price = [] #
line_disc = [] #
container_scheme = []
Gst = []
extended_price = [] #
price_per_iteam = [] #

product_description = table_df['Unnamed: 0'].tolist()
line_disc = table_df['Unnamed: 2'].tolist()
extended_price = table_df['Document No: 223163058'].tolist()
price_per_iteam = table_df['Unnamed: 8'].tolist()
list_PQL = table_df['P.O. Box 3012'].tolist()

for i in list_PQL:
    j=i.split(" ")
    product_code.append(j[0])
    quantity.append(j[2:4])
    list_price.append(j[4])


list_Con_GST = table_df['Unnamed: 4'].tolist()
new_con_GST = []
for iteam in list_Con_GST:
    new_con_GST.append(str(iteam))

for i in new_con_GST:
    j = i.split(" ")
    if len(j)==0:
        container_scheme.append('nan')
        Gst.append('nan')
    elif len(j)==1:
        container_scheme.append(j[0])
        Gst.append('nan')
    elif len(j)==2:
        container_scheme.append(j[0])
        Gst.append(j[1])
        
data_dic = {'Product Description':product_description,'Product Code':product_code,'Quantity':quantity,'List Price':list_price,
            'Line Disc':line_disc,'Container Scheme':container_scheme,'GST':Gst,'Extended Price':extended_price,
            'Price per Item':price_per_iteam}


dataframe = pd.DataFrame(data=data_dic)
dataframe = dataframe.append({'Extended Price':pd.to_numeric(dataframe['Extended Price']).sum()},ignore_index=True)
dataframe.to_csv(filename+'.csv',index=None,header=True)
```
